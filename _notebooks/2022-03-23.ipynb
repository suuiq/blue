# 과제 2.6

```python
bn=bx1.groupby(bx1.age).mean().reset_index()
#연령별 평균
```

```python
bn
```

```python
from scipy import stats
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('id ~ age',bn).fit()
anova_lm(model)
```

```python
model = ols('year ~ age',bn).fit()
anova_lm(model)
```

```python
model = ols('buy ~ age',bn).fit()
anova_lm(model)
```

```python
model = ols('edu ~ age',bn).fit()
anova_lm(model)
```

```python
model = ols('income ~ age',bn).fit()
anova_lm(model)
```

```python
model = ols('exp ~ age',bn).fit()
anova_lm(model)
```

```python
model = ols('co ~ age',bn).fit()
anova_lm(model)
```

```python
model = ols('ad ~ age',bn).fit()
anova_lm(model)
```

```python
model = ols('prod ~ age',bn).fit()
anova_lm(model)
```

p-값이 0.05보다 작으면 유의미한 차이가 있다. 분산분석 실행 결과, ad와 age 사이의 p-값만 0.05보다 작으므로 유의미한 것을 알 수 있다.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model = smf.ols(formula = 'income ~ id + year + buy + age + edu + exp + co + ad + prod', data = bx1)
result = model.fit()
result.summary()
```

```python
import statsmodels.api as sm
logis = sm.Logit.from_formula('buy ~ id+year+age+edu+income+exp+co+ad+prod', bx1).fit()
logis.summary()
```

P>|z| 부분을 보면 유의확률을 알 수 있다, 독립변수의 유의확률이 0.05보다 작으면 독립변수가 종속변수에 영향을 미치는 것이 유의미하다고 본다. 
따라서 edu, exp, co, ad는 buy에 유의미하게 영향을 미친다.
